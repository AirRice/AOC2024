```python
import pandas as pd
import numpy as np
import random
```

### Day 1

```python
input1 = open('inputs/day1_1', 'r')
input1_lines = input1.readlines()
input_array = np.empty([2, len(input1_lines)], dtype = int) #Make a new array of the length
for idx, line in enumerate(input1_lines):
    input_array[:, idx] = np.array(line.split(), dtype = int)
input_array_sorted = np.rint(np.sort(input_array))
sum = 0
for numpair in input_array_sorted.T:
    sum += abs(numpair[0] - numpair[1])
print(sum)
```

Day 1 part 2

```python
sum2 = 0
for input_num in input_array_sorted[0]:
    counts = np.count_nonzero(input_array_sorted[1] == input_num)
    sum2 += input_num * counts
    print(str(input_num) + " * " + str(counts) + " = " + str(input_num * counts))
    print(sum2)
print(sum2)
```

### Day 2

```python
input2 = open('inputs/day2_1', 'r')
input2_lines = input2.readlines()
safe_sum = 0
for idx, line in enumerate(input2_lines):
    is_safe = True
    line_split = np.array(line.split(), dtype = int)
    print(line_split)
    is_safe = (np.array_equal(np.sort(line_split), line_split) or np.array_equal(np.flip(np.sort(line_split)), line_split))
    if (is_safe):
        for i in range(1, line_split.shape[0]):
            i_diff = abs(line_split[i-1] - line_split[i])
            if (i_diff < 1 or i_diff > 3):
                print("UNSAFE : Difference is " + str(i_diff) + " For elements " + str(line_split[i-1]) + ", " + str(line_split[i]))
                is_safe = False
                break
    if (is_safe):
        safe_sum += 1
print(safe_sum)
```

Day 2 part 2 (more robust algorithm)

```python
def check_elements_safe(a, b, line_split, sort_type):
    if (a < 0 or b >= line_split.shape[0]):
        return True
    sort_type_cur = 0 if (line_split[a] < line_split[b]) else 1
    diff = abs(line_split[a] - line_split[b])
    print("elements " + str(line_split[a]) + ", " + str(line_split[b]))
    return sort_type == sort_type_cur and diff >= 1 and diff <= 3
def check_line_split_safe(line_split, sort_type):
    print(line_split)
    bad_idx = []
    for i in range(0, line_split.shape[0]-1):
        if i in bad_idx:
            continue
        if not check_elements_safe(i, (i+1 if not i+1 in bad_idx else i+2), line_split, sort_type):
            print("Bad elements " + str(line_split[i]) + ", " + str(line_split[i+1]))
            if check_elements_safe(i, (i+2 if not i+2 in bad_idx else i+3), line_split, sort_type):
                bad_idx.append(i+1)
            elif check_elements_safe((i-1 if not i-1 in bad_idx else i-2), (i+1 if not i+1 in bad_idx else i+2), line_split, sort_type):
                bad_idx.append(i)
            else:
                print("UNSAFE!!")
                return False
    return len(bad_idx) <= 1

input2 = open('inputs/day2_1', 'r')
input2_lines = input2.readlines()
safe_sum = 0


for idx, line in enumerate(input2_lines):
    line_split = np.array(line.split(), dtype = int)
    if (check_line_split_safe(line_split, 0) or check_line_split_safe(line_split, 1)):
        safe_sum += 1
print(safe_sum)
```
